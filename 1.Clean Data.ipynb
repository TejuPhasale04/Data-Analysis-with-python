## Filling Missing values

## 1)Numerical data-use mean

```python
df3=df.copy()
```

```python
df3.head(3)
```

```python
mean_val=df3["Engine HP"].mean()
df3["Engine HP"].fillna(mean_val,inplace=True)
```

```python
df3.isna().sum()
```

## 2)Categorial Data

```python
df4=df.copy()
```

```python
df4.isna().sum()
```

```python
df4["Market Category"].fillna("No Category" ,inplace=True)
```

```python
df4.isna().sum()
```

```python
df4.head(3)
```

```python
df4["Make"].value_counts()
```
